# Example Jupyter Notebook
To suppress the TensorFlow backend warnings and info logs (like the NUMA node messages and factory registration errors), you can set the `TF_CPP_MIN_LOG_LEVEL` environment variable **before** importing TensorFlow.

### Log Levels Breakdown
- **`0`**: Default (all logs shown).
- **`1`**: Filters out `INFO` logs.
- **`2`**: Filters out `INFO` and `WARNING` logs.
- **`3`**: Filters out `INFO`, `WARNING`, and `ERROR` logs.

Setting it to **`3`** is the most effective for cleaning up the notebook output completely, as it will suppress the "Unable to register factory" errors and the NUMA node info messages you are seeing.

```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'  # 0=all, 1=INFO, 2=WARNING, 3=ERROR suppressed
import tensorflow as tf

print(f"TensorFlow Version: {tf.__version__}")
gpus = tf.config.list_physical_devices('GPU')

if gpus:
    print(f"Found {len(gpus)} GPU(s):")
    for gpu in gpus:
        details = tf.config.experimental.get_device_details(gpu)
        name = details.get('device_name', 'Unknown Device')
        print(f"- {name} ({gpu.name})")
else:
    print("No GPU detected.")
```
